```python
# Diagramme en bar de sebder1
plt.rcParams["figure.figsize"]=[16,9]
data=df["Sender1"].value_counts()

plt.bar(data.index, data, color=["blue", "red"])
plt.xlabel("Pays santionant")
plt.ylabel("Nombre de sanctions")
plt.title("Nombre de santions selon les pays")
plt.xticks(rotation="vertical")
plt.draw()
plt.show()
```
